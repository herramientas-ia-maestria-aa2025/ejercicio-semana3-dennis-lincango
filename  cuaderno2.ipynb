```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cuaderno1.csv')
```

```python
df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupación y estadísticas
resultado = df.groupby('edad').agg(
    valor_count_edad=('edad', 'size')  
).reset_index()

# Crear una figura y ejes para los gráficos
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: número de registros por cada edad
sns.barplot(x='edad', y='valor_count_edad', hue='edad', data=resultado, palette='deep', ax=axs[0])
axs[0].set_title('Número de estudiantes por grupo de edad')
axs[0].set_ylabel('Cantidad de estudiantes')

# Gráfico 2: número de registros por cada provincia
resultado_provincia = df.groupby('provincia').agg(
    valor_count_provincia=('provincia', 'size')  
).reset_index()

sns.barplot(x='provincia', y='valor_count_provincia', hue='provincia', data=resultado_provincia, palette='muted', ax=axs[1])
axs[1].set_title('Número de registros por provincia')  # Corregir el título
axs[1].set_ylabel('Cantidad de registros')

# Mostrar los gráficos
# plt.tight_layout()
plt.show()
```
